```python
import tensorflow as tf
```

```python
daily_sales_number = [20, 23, -23, 34, -44, 56, 89, 90]

dataset = tf.data.Dataset.from_tensor_slices(daily_sales_number)

```

```python
for sales in dataset.as_numpy_iterator():
    # print(sales.numpy())
    print(sales)
```

```python
for sales in dataset.take(3):
    print(sales)
```

```python
dataset = dataset.filter(lambda x: x>0)

for sales in dataset.take(-1):
    print(sales)
```

```python
dataset = dataset.map(lambda x: x*10)
for sales in dataset.as_numpy_iterator():
    print(sales)
```

```python
dataset = dataset.shuffle(3)
for sales in dataset.as_numpy_iterator():
    print(sales)
```

```python
for sales_batch in dataset.batch(3):
    print(sales_batch.numpy())
```
