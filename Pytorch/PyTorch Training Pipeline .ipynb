```python
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv('data/breast-cancer.csv')
df.head()
```

```python
print(df.shape)
print(type(df))
print(df.info())
print(df.isnull().sum())
```

```python
df = df.drop(columns=['id'])
df.head()
```

```python
y = df['diagnosis'].map({'M': 1, 'B': 0})
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['diagnosis']), y, test_size=0.2, random_state=42)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
print()
print(y.value_counts())
```

```python
b_count = 0
m_count = 0
for i in y:
    if i == 0:
        b_count += 1
    else:
        m_count += 1

print(f'Benign count: {b_count}')
print(f'Malignant count: {m_count}')
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print(X_train_scaled)
print(X_test_scaled)
```

```python
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
print(y_train_encoded)
print(y_test_encoded)
```

**numpy arrays to pytorch tensors**

```python
X_train_tensor = torch.from_numpy(X_train_scaled)
X_test_tensor = torch.from_numpy(X_test_scaled)
y_train_tensor = torch.from_numpy(y_train_encoded)
y_test_tensor = torch.from_numpy(y_test_encoded)
print(X_train_tensor)
print(X_test_tensor)
print(y_train_tensor)
print(y_test_tensor)
```

```python
from torch.utils.data import Dataset, DataLoader
class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels
    
    def __len__(self):
        return self.features.shape[0]
    
    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
train_dataset = CustomDataset(X_train_tensor, y_train_tensor)
test_dataset = CustomDataset(X_test_tensor, y_test_tensor)
print(len(train_dataset))
print(len(test_dataset))
```

```python
# class MySimpleNN():
#     def __init__(self, X):
#         self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
#         self.bias = torch.rand(1, dtype=torch.float64, requires_grad=True)

#     def forward(self, X):
#         return torch.sigmoid(X @ self.weights + self.bias)
    
#     def loss_function(self, y_pred, y_true):
#         epsilon = 1e-7
#         y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
#         loss = -torch.mean(y_true * torch.log(y_pred) + (1 - y_true) * torch.log(1 - y_pred))
#         return loss
    

import torch.nn as nn
class MySimpleNN(nn.Module):
    def __init__(self, X):
        super().__init__()
        # self.networks = nn.Sequential(
        #     nn.Linear(X.shape[1], 1),
        #     nn.Sigmoid()
        # )
        self.linear = nn.Linear(X.shape[1], 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, X):
        return self.sigmoid(self.linear(X))
    
```

```python
learning_rate = 0.01
epochs = 30
#loss
loss_function = nn.BCELoss()
#optimizer
model = MySimpleNN(X_train_tensor.float())
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
# model = MySimpleNN(X_train_tensor)
# for epoch in range(epochs):
#     # Forward pass
#     y_pred = model.forward(X_train_tensor)
#     # print(y_pred)

#     # Compute loss
#     loss = model.loss_function(y_pred, y_train_tensor)
#     # print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item()}")

#     # Backward pass
#     loss.backward()

#     # Update weights and bias
#     with torch.no_grad():
#         model.weights -= learning_rate * model.weights.grad
#         model.bias -= learning_rate * model.bias.grad

#     # Zero gradients
#     model.weights.grad.zero_()
#     model.bias.grad.zero_()

#     print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item()}")

# OR

for epoch in range(epochs):

    for batch_features, batch_labels in DataLoader(train_dataset, batch_size=2, shuffle=True):
        # Forward pass
        y_pred = model(batch_features.float())

        # Compute loss
        loss = loss_function(y_pred, batch_labels.view(-1, 1).float())
        # print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item()}")
        
        # Zero gradients(must do before backward pass recommended by PyTorch)
        optimizer.zero_grad()
        
        # Backward pass
        loss.backward()

        # Update weights and bias
        optimizer.step()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item()}")

```

```python
print(model.linear.weight)
print(model.linear.bias)
```

```python
model.eval()
accuracy_list = []

with torch.no_grad():
    for batch_features, batch_labels in DataLoader(test_dataset, batch_size=2, shuffle=False):
        y_test_pred = model(batch_features.float())
        y_test_pred_labels = (y_test_pred >= 0.5).float()
        accuracy = (y_test_pred_labels.squeeze() == batch_labels).float().mean()
        accuracy_list.append(accuracy.item())

print(f"Test Accuracy: {sum(accuracy_list) / len(accuracy_list) * 100:.2f}%")
```
