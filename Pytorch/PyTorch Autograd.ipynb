```python
import torch

x = torch.tensor(6.7)
y = torch.tensor(0.0)

w = torch.tensor(1.0)
b = torch.tensor(0.0)
```

```python
#loss = binary_cross_entropy(prediction, target)
def binary_cross_entropy(prediction, target):
    epsilon = 1e-8
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return - (target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
```

```python
#forward pass
z = w * x + b
y_pred = torch.sigmoid(z)

loss = binary_cross_entropy(y_pred, y)
print(loss)
```

```python
#derivatives
dloss_dy_pred = (y_pred - y) / (y_pred * (1 - y_pred))

dy_pred_dz = y_pred * (1 - y_pred)
dz_dw = x
dz_db = 1.0

dloss_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dloss_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f'manual dloss_dw: {dloss_dw.item()}')
print(f'manual dloss_db: {dloss_db.item()}')
```

**using Autograd**

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)

w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

z = w * x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy(y_pred, y)
loss.backward()
print(f'autograd dloss_dw: {w.grad.item()}')
print(f'autograd dloss_db: {b.grad.item()}')
```

```python
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
y = x ** 2
z = y.sum()
z.backward()
print(x.grad)
```

```python
x.requires_grad_(False)
print(x.grad)
```

```python
x = torch.ones(2, 2, requires_grad=True)
y = x + 2
z = y * y * 3
out = z.mean()
print(out) 
out.backward()
print(x.grad) 
```
